```python
import folium

m = folium.Map(
    location=[37.5838699,127.0565831],
    zoom_start=10
)

import json
with open('./static/data/test.json',mode='rt',encoding='utf-8') as f:
    geo = json.loads(f.read())
    f.close()

folium.GeoJson(
    geo,
    name='SIG_KOR_NM'
).add_to(m)

m.save('map.html')
```
